# Define the function to use in Django website

```python
# Load data
df_data = pd.read_csv('cna_news_topkey_with_category_via_token_pos.csv',sep=',')

# Prepare top keys data with dictionary format
data={}
for idx, row in df_data.iterrows():
    data[row['category']]= eval(row['top_keys'])    
```

```python
data
```

```python
# Here is the function. We will use it for drawing in django website
def get_category_topword(cate, topk=10):
    wf_pairs = data[cate][0:topk]
    words = [w for w, f in wf_pairs]
    freqs = [f for w, f in wf_pairs]
    chart_data = {
        "category": cate,
        "labels": words,
        "values": freqs}
    return chart_data, wf_pairs
```

```python
get_category_topword('學術股市',5)
```
