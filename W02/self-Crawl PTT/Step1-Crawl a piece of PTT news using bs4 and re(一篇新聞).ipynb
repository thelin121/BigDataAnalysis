## using try except to set empty photo_link if there is no photo link

```python
try:
    photo_link = page.find('figure',{'class':"floatImg center"}).find('img').get('src')
except:
    photo_link = ''
```

```python
img_src = page.find('figure',{'class':"floatImg center"}).find('img')
try:
    if img_src.has_attr('data-src'):
        photo_link = img_src.get('data-src')
    else:
        photo_link = img_src.get('src')
except:
    photo_link = ''
print(photo_link)
```

```python

content = page.find('section',{'class':"layout__top layout-with-rail__top"})

# Remove unnecessary content

# Remove "透過 Google News追蹤中央社"
try:
    content.find('div', {'class':"gmailNews"}).decompose()
except:
    pass

# remove "訂閱早安電子報"
try:
    content.find('div', {'class':"subscriInnTitle"}).decompose()
except:
    pass

# 額外插入的訂閱連結-->移除之
try:
    content.find('div', {'class':"SubscriptionInner mySubscriptionInner"}).decompose()
except:
    pass

# 額外插入的延伸閱讀區塊-->移除之
try:
    content.find('div', {'class':"paragraph moreArticle"}).decompose()
except:
    pass

# 額外插入的圖片與文字說明區塊-->移除之
try:
    for media in content.findAll('div', {'class':"media"}):
        media.decompose()
except:
    pass
```

# Remove some unnecessary words in content using re

## remove （中央社記者鍾佑貞台北13日電）

```python
content=page.find('div',{'class':"paragraph"}).text
```

```python
content
```

```python
# remove （中央社記者鍾佑貞台北13日電）
# 試圖找到第一個右括號 ?表示非貪婪模式 +表示一個或多個
text = re.sub('（中央社記者.+?）','', content)
text
```

```python
# remove （中央社記者鍾佑貞台北13日電）
# 試圖找到第一個右括號 ?表示非貪婪模式 *表示0個或多個
text = re.sub('（中央社記者.*?）','', content)
text
```

```python
# 這樣失敗: 試圖找到最後一個右括號
text = re.sub('（中央社記者.*）','', content) 
text
```

### Remove （編輯：郭無患）1100313....

```python
# How can I delete rest of the line after a specific string?
# （編輯：郭無患）1100313延伸閱讀傳自製30公厘訓練彈品質不佳 國防部：符原廠要求雲豹甲車奔馳165公里驗證性能 學童揮國旗喊加油國軍戰車隱匿廢五金堆蓄勢待發 創意源自變形金剛'
text = re.sub('（編輯.*','', content) 
text
```

```python
# 這樣失敗: 因為遇到數字或其他非文字就停止了
text = re.sub('（編輯：\w*','', content)
text
```
