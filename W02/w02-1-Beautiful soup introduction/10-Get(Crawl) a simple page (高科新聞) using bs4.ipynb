# Crawl a simple page using beautifulsoup4

爬取一篇簡單的頁面的內容:使用bs4

自己動手抓不難喔!

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://www.nkust.edu.tw/p/406-1000-41300,r12.php?Lang=zh-tw'
r = requests.get( url )
```

```python
r
```

```python
print(r.text)
```

```python
page = BeautifulSoup(r.text,'lxml')
```

```python
page
```

# Using find

只會回傳第一個符合條件的元素

```python
page.find(name = "p")
```

```python
page.find(name = "p", style="text-align: justify;")
```

```python
# the same result
page.find(name = "p", attrs = {'style' : "text-align: justify;"})
```

```python
# get content text
page.find(name = "p", attrs = {'style' : "text-align: justify;"}).text
```

# Using findAll

```python
page.findAll('p', style="text-align: justify;")
```

```python
[content.text for content in page.findAll('p', style="text-align: justify;")]
```

```python
nkust_news = [content.text for content in page.findAll('p', style="text-align: justify;")]
```

```python
nkust_news
```

```python
"\n".join(nkust_news)
```

```python
result = "\n".join(nkust_news)
```

```python
print(result)
```

## Best way to crawl photo

```python
page.find('img', {"class": "img-responsive"})
```

```python
page.findAll('img', {"class": "img-responsive"})
```

```python
for img in page.findAll('img', {"class": "img-responsive"}):
    print(img['src'])
```

```python
[img['src'] for img in page.findAll('img', {"class": "img-responsive"})]
```
